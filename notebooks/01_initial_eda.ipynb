```python
# Keep only the columns you'll actually use for analysis
df = df[[
    'Prscrbr_NPI',
    'Prscrbr_State_Abrvtn',
    'Prscrbr_Type',
    'Brnd_Name',
    'Gnrc_Name',
    'Tot_Clms',
    'Tot_Drug_Cst',
    'Tot_Benes'
]]

print("After cleanup:")
print(f"Rows: {len(df):,}")
print(f"Columns: {df.shape[1]}")
print(df.columns.tolist())
```

```python
df.describe()
```

```python
unique_prescribers = df['Prscrbr_NPI'].nunique()
total_rows = len(df)

print(f"Total rows: {total_rows:,}")
print(f"Unique prescribers: {unique_prescribers:,}")
print(f"Avg brands per prescriber: {total_rows / unique_prescribers:.2f}")
```

```python
# How many brands does each prescriber use?
brands_per_doctor = df.groupby('Prscrbr_NPI')['Brnd_Name'].nunique()

# Distribution
print("Number of doctors prescribing X different brands:")
print(brands_per_doctor.value_counts().sort_index())

print("\nAs percentages:")
print((brands_per_doctor.value_counts(normalize=True).sort_index() * 100).round(1))
```

```python
# Brand-level summary
brand_summary = df.groupby('Brnd_Name').agg(
    total_claims=('Tot_Clms', 'sum'),
    total_cost=('Tot_Drug_Cst', 'sum'),
    unique_prescribers=('Prscrbr_NPI', 'nunique'),
    avg_claims_per_prescriber=('Tot_Clms', 'mean')
).sort_values('total_claims', ascending=False)

# Add a market share column (% of total claims)
brand_summary['market_share_pct'] = (
    brand_summary['total_claims'] / brand_summary['total_claims'].sum() * 100
).round(1)

# Add cost per claim
brand_summary['avg_cost_per_claim'] = (
    brand_summary['total_cost'] / brand_summary['total_claims']
).round(2)

print(brand_summary)
```

```python
specialty_summary = df.groupby('Prscrbr_Type').agg(
    total_claims=('Tot_Clms', 'sum'),
    unique_prescribers=('Prscrbr_NPI', 'nunique')
).sort_values('total_claims', ascending=False).head(15)

specialty_summary['claims_share_pct'] = (
    specialty_summary['total_claims'] / specialty_summary['total_claims'].sum() * 100
).round(1)

print(specialty_summary)
```

```python
state_summary = df.groupby('Prscrbr_State_Abrvtn').agg(
    total_claims=('Tot_Clms', 'sum'),
    unique_prescribers=('Prscrbr_NPI', 'nunique')
).sort_values('total_claims', ascending=False).head(15)

state_summary['claims_share_pct'] = (
    state_summary['total_claims'] / state_summary['total_claims'].sum() * 100
).round(1)

print(state_summary)
```
