```python
df_2022 = pd.read_csv("C:/Users/ADMIN/Desktop/glp1-project/data/glp1_medicare_2022.csv", low_memory=False)

print("2022 loaded successfully!")
print(f"Rows: {len(df_2022):,}")
print(f"Columns: {df_2022.shape[1]}")
print(f"Unique prescribers: {df_2022['Prscrbr_NPI'].nunique():,}")
print(f"\nBrand breakdown:")
print(df_2022['Brnd_Name'].value_counts())
```

```python
# ── 2022 brand shares ──────────────────────────────────────
brand_2022 = df_2022.groupby('Brnd_Name').agg(
    total_claims_2022=('Tot_Clms', 'sum'),
    prescribers_2022=('Prscrbr_NPI', 'nunique')
)
brand_2022['share_2022'] = (
    brand_2022['total_claims_2022'] /
    brand_2022['total_claims_2022'].sum() * 100
).round(1)

# ── 2023 brand shares ──────────────────────────────────────
brand_2023 = df.groupby('Brnd_Name').agg(
    total_claims_2023=('Tot_Clms', 'sum'),
    prescribers_2023=('Prscrbr_NPI', 'nunique')
)
brand_2023['share_2023'] = (
    brand_2023['total_claims_2023'] /
    brand_2023['total_claims_2023'].sum() * 100
).round(1)

# ── Combine into one table ─────────────────────────────────
comparison = brand_2022[['share_2022', 'total_claims_2022', 'prescribers_2022']].join(
    brand_2023[['share_2023', 'total_claims_2023', 'prescribers_2023']],
    how='outer'
).fillna(0)

# ── Add change columns ─────────────────────────────────────
comparison['share_change'] = (
    comparison['share_2023'] - comparison['share_2022']
).round(1)

comparison['prescriber_growth'] = (
    comparison['prescribers_2023'] - comparison['prescribers_2022']
).astype(int)

comparison = comparison.sort_values('share_2023', ascending=False)

# ── Print brand level table ────────────────────────────────
print("=" * 65)
print("BRAND LEVEL: YEAR-OVER-YEAR SHARE SHIFT 2022 → 2023")
print("=" * 65)
print(comparison[[
    'share_2022',
    'share_2023',
    'share_change',
    'prescribers_2022',
    'prescribers_2023',
    'prescriber_growth'
]])

# ── Print company level totals ─────────────────────────────
print("\n" + "=" * 65)
print("COMPANY LEVEL: NOVO vs LILLY")
print("=" * 65)

novo_2022 = comparison.loc[['Ozempic','Rybelsus'], 'share_2022'].sum().round(1)
novo_2023 = comparison.loc[['Ozempic','Rybelsus'], 'share_2023'].sum().round(1)
lilly_2022 = comparison.loc[['Trulicity','Mounjaro'], 'share_2022'].sum().round(1)
lilly_2023 = comparison.loc[['Trulicity','Mounjaro'], 'share_2023'].sum().round(1)

print(f"Novo  2022: {novo_2022}%  →  2023: {novo_2023}%  (change: {round(novo_2023-novo_2022,1):+}pp)")
print(f"Lilly 2022: {lilly_2022}%  →  2023: {lilly_2023}%  (change: {round(lilly_2023-lilly_2022,1):+}pp)")

# ── The hypothesis test ────────────────────────────────────
print("\n" + "=" * 65)
print("HYPOTHESIS TEST")
print("=" * 65)

mounjaro_gain = comparison.loc['Mounjaro', 'share_change']
ozempic_change = comparison.loc['Ozempic', 'share_change']
trulicity_change = comparison.loc['Trulicity', 'share_change']

print(f"Mounjaro gained:    {mounjaro_gain:+.1f} percentage points")
print(f"Ozempic changed:    {ozempic_change:+.1f} percentage points")
print(f"Trulicity changed:  {trulicity_change:+.1f} percentage points")
print()

if abs(ozempic_change) > abs(trulicity_change):
    print("VERDICT: Mounjaro drew more share from OZEMPIC than Trulicity")
    print("→ Hypothesis CONFIRMED: Direct threat to Novo's franchise")
elif abs(trulicity_change) > abs(ozempic_change):
    print("VERDICT: Mounjaro drew more share from TRULICITY than Ozempic")
    print("→ Hypothesis WRONG: Primarily Lilly self-cannibalization")
else:
    print("VERDICT: Mixed — Mounjaro drew equally from both brands")
    print("→ Hypothesis PARTIALLY confirmed")
```

```python
# ─────────────────────────────────────────────────────────────
# STEP 1: Build deciles for BOTH years separately
# ─────────────────────────────────────────────────────────────

# 2022 deciles
prescriber_volume_2022 = df_2022.groupby('Prscrbr_NPI')['Tot_Clms'].sum().reset_index()
prescriber_volume_2022.columns = ['Prscrbr_NPI', 'total_glp1_claims']
prescriber_volume_2022['decile'] = pd.qcut(
    prescriber_volume_2022['total_glp1_claims'],
    q=10,
    labels=range(1, 11),
    duplicates='drop'
)

# 2023 deciles (you already built this — recreating for clarity)
prescriber_volume_2023 = df.groupby('Prscrbr_NPI')['Tot_Clms'].sum().reset_index()
prescriber_volume_2023.columns = ['Prscrbr_NPI', 'total_glp1_claims']
prescriber_volume_2023['decile'] = pd.qcut(
    prescriber_volume_2023['total_glp1_claims'],
    q=10,
    labels=range(1, 11),
    duplicates='drop'
)

# ─────────────────────────────────────────────────────────────
# STEP 2: Add company labels
# ─────────────────────────────────────────────────────────────
def assign_company(brand):
    if brand in ['Ozempic', 'Rybelsus']:
        return 'Novo'
    elif brand in ['Trulicity', 'Mounjaro']:
        return 'Lilly'
    else:
        return 'Other'

df['Company'] = df['Brnd_Name'].apply(assign_company)
df_2022['Company'] = df_2022['Brnd_Name'].apply(assign_company)

# ─────────────────────────────────────────────────────────────
# STEP 3: Merge decile labels into both dataframes
# ─────────────────────────────────────────────────────────────
df_2022_d = df_2022.merge(
    prescriber_volume_2022[['Prscrbr_NPI', 'decile']],
    on='Prscrbr_NPI',
    how='left'
)

df_2023_d = df.merge(
    prescriber_volume_2023[['Prscrbr_NPI', 'decile']],
    on='Prscrbr_NPI',
    how='left'
)

# ─────────────────────────────────────────────────────────────
# STEP 4: Compute Novo share per decile, both years
# ─────────────────────────────────────────────────────────────
def company_share_by_decile(dframe, year_label):
    by_decile_company = dframe.groupby(
        ['decile', 'Company'],
        observed=True
    )['Tot_Clms'].sum().reset_index()
    
    pivoted = by_decile_company.pivot(
        index='decile',
        columns='Company',
        values='Tot_Clms'
    ).fillna(0)
    
    pivoted['total'] = pivoted['Novo'] + pivoted['Lilly']
    pivoted[f'novo_share_{year_label}'] = (pivoted['Novo'] / pivoted['total'] * 100).round(1)
    pivoted[f'lilly_share_{year_label}'] = (pivoted['Lilly'] / pivoted['total'] * 100).round(1)
    
    return pivoted[[f'novo_share_{year_label}', f'lilly_share_{year_label}']]

share_2022 = company_share_by_decile(df_2022_d, '2022')
share_2023 = company_share_by_decile(df_2023_d, '2023')

# ─────────────────────────────────────────────────────────────
# STEP 5: Build the comparison table
# ─────────────────────────────────────────────────────────────
decile_company_comparison = share_2022.join(share_2023)

decile_company_comparison['novo_share_change'] = (
    decile_company_comparison['novo_share_2023'] -
    decile_company_comparison['novo_share_2022']
).round(1)

decile_company_comparison['lilly_share_change'] = (
    decile_company_comparison['lilly_share_2023'] -
    decile_company_comparison['lilly_share_2022']
).round(1)

print("=" * 75)
print("NOVO vs LILLY: SHARE BY DECILE, 2022 → 2023")
print("=" * 75)
print(decile_company_comparison.to_string())

# ─────────────────────────────────────────────────────────────
# STEP 6: Headline read
# ─────────────────────────────────────────────────────────────
print("\n" + "=" * 75)
print("READ: WHERE DID NOVO WIN, AND BY HOW MUCH?")
print("=" * 75)

for decile in range(1, 11):
    change = decile_company_comparison.loc[decile, 'novo_share_change']
    if change > 0:
        verdict = "Novo gained"
    elif change < 0:
        verdict = "Novo LOST"
    else:
        verdict = "No change"
    print(f"  Decile {decile:2d}: {verdict} {abs(change)}pp")
```

```python
# ─────────────────────────────────────────────────────────────
# BULLET 3: Brand share within each decile (2023)
# ─────────────────────────────────────────────────────────────

# Sum claims per (decile, brand)
decile_brand = df_2023_d.groupby(
    ['decile', 'Brnd_Name'],
    observed=True
)['Tot_Clms'].sum().reset_index()

# Total claims per decile (denominator for within-decile share)
decile_brand['decile_total'] = decile_brand.groupby(
    'decile',
    observed=True
)['Tot_Clms'].transform('sum')

# Within-decile brand share
decile_brand['brand_share_pct'] = (
    decile_brand['Tot_Clms'] /
    decile_brand['decile_total'] * 100
).round(1)

# Pivot for readability
decile_pivot = decile_brand.pivot(
    index='decile',
    columns='Brnd_Name',
    values='brand_share_pct'
).fillna(0)

decile_pivot = decile_pivot[['Ozempic', 'Trulicity', 'Mounjaro', 'Rybelsus']]

print("BRAND SHARE WITHIN EACH DECILE — 2023")
print("=" * 65)
print(decile_pivot.to_string())
```

```python
# ─────────────────────────────────────────────────────────────
# BULLET 4: Decile structure within each major specialty (2023)
# ─────────────────────────────────────────────────────────────

# Get per-prescriber total volume + their primary specialty
# (Each doctor has one specialty in this data)
prescriber_with_specialty = df.groupby(
    ['Prscrbr_NPI', 'Prscrbr_Type']
)['Tot_Clms'].sum().reset_index()
prescriber_with_specialty.columns = ['Prscrbr_NPI', 'Specialty', 'total_glp1_claims']

# Focus on the top 5 specialties (96% of volume)
top_5_specialties = [
    'Family Practice',
    'Internal Medicine',
    'Nurse Practitioner',
    'Endocrinology',
    'Physician Assistant'
]

prescriber_top5 = prescriber_with_specialty[
    prescriber_with_specialty['Specialty'].isin(top_5_specialties)
].copy()

# Build separate deciles within each specialty
prescriber_top5['within_specialty_decile'] = prescriber_top5.groupby(
    'Specialty'
)['total_glp1_claims'].transform(
    lambda x: pd.qcut(x, q=10, labels=range(1, 11), duplicates='drop')
)

# Show decile cutoffs per specialty
print("DECILE CUTOFFS BY SPECIALTY (2023)")
print("=" * 75)

for spec in top_5_specialties:
    spec_data = prescriber_top5[prescriber_top5['Specialty'] == spec]
    
    decile_stats = spec_data.groupby('within_specialty_decile', observed=True).agg(
        prescribers=('Prscrbr_NPI', 'count'),
        min_claims=('total_glp1_claims', 'min'),
        max_claims=('total_glp1_claims', 'max'),
        avg_claims=('total_glp1_claims', 'mean')
    ).round(1)
    
    print(f"\n{spec}")
    print("-" * 75)
    print(decile_stats.to_string())
```

```python
# ─────────────────────────────────────────────────────────────
# PRESSURE TEST 1: Build 2022 brand-by-decile pivot for comparison
# ─────────────────────────────────────────────────────────────

# Step 1a: Sum claims per (decile, brand) for 2022
decile_brand_2022 = df_2022_d.groupby(
    ['decile', 'Brnd_Name'],
    observed=True
)['Tot_Clms'].sum().reset_index()

# Step 1b: Total claims per decile (denominator for within-decile share)
decile_brand_2022['decile_total'] = decile_brand_2022.groupby(
    'decile',
    observed=True
)['Tot_Clms'].transform('sum')

# Step 1c: Within-decile brand share
decile_brand_2022['brand_share_pct'] = (
    decile_brand_2022['Tot_Clms'] /
    decile_brand_2022['decile_total'] * 100
).round(1)

# Step 1d: Pivot for readability
decile_pivot_2022 = decile_brand_2022.pivot(
    index='decile',
    columns='Brnd_Name',
    values='brand_share_pct'
).fillna(0)

# Reorder columns to match 2023 layout
decile_pivot_2022 = decile_pivot_2022[['Ozempic', 'Trulicity', 'Mounjaro', 'Rybelsus']]

print("BRAND SHARE WITHIN EACH DECILE — 2022")
print("=" * 65)
print(decile_pivot_2022.to_string())
```

```python
# ─────────────────────────────────────────────────────────────
# YoY change in brand share within each decile
# ─────────────────────────────────────────────────────────────

# Combine the two pivots
yoy_change = decile_pivot - decile_pivot_2022

print("YOY CHANGE IN BRAND SHARE WITHIN EACH DECILE (2022 → 2023)")
print("=" * 70)
print(yoy_change.round(1).to_string())
print()
print("Positive number = brand gained share within that decile")
print("Negative number = brand lost share within that decile")
```

```python
# ─────────────────────────────────────────────────────────────
# PRESSURE TEST 1: Sample size check
# Identify any specialty × decile cell with fewer than 100 prescribers
# ─────────────────────────────────────────────────────────────

# Recompute within-specialty prescriber counts (you've done this before)
top_5_specialties = [
    'Family Practice', 'Internal Medicine', 'Nurse Practitioner',
    'Endocrinology', 'Physician Assistant'
]

prescriber_with_specialty = df.groupby(
    ['Prscrbr_NPI', 'Prscrbr_Type']
)['Tot_Clms'].sum().reset_index()
prescriber_with_specialty.columns = ['Prscrbr_NPI', 'Specialty', 'total_glp1_claims']

prescriber_top5 = prescriber_with_specialty[
    prescriber_with_specialty['Specialty'].isin(top_5_specialties)
].copy()

prescriber_top5['within_specialty_decile'] = prescriber_top5.groupby(
    'Specialty'
)['total_glp1_claims'].transform(
    lambda x: pd.qcut(x, q=10, labels=range(1, 11), duplicates='drop')
)

# Count prescribers in each (specialty, decile) cell
sample_size_check = prescriber_top5.groupby(
    ['Specialty', 'within_specialty_decile'],
    observed=True
).size().reset_index(name='prescriber_count')

# Pivot for readability
sample_pivot = sample_size_check.pivot(
    index='within_specialty_decile',
    columns='Specialty',
    values='prescriber_count'
).fillna(0).astype(int)

# Reorder columns
sample_pivot = sample_pivot[top_5_specialties]

print("PRESCRIBER COUNT BY (SPECIALTY x WITHIN-SPECIALTY DECILE)")
print("=" * 70)
print(sample_pivot.to_string())

# Flag cells with <100 prescribers
print("\n" + "=" * 70)
print("CELLS WITH <100 PRESCRIBERS (potential sample size concerns)")
print("=" * 70)
small_cells = sample_pivot[sample_pivot < 100]
print(small_cells.fillna('OK').to_string())
```

```python
# ─────────────────────────────────────────────────────────────
# PRESSURE TEST 2: Re-run YoY analysis with QUINTILES (5 buckets)
# Tests whether the split-pyramid pattern is bucketing-dependent
# ─────────────────────────────────────────────────────────────

# Step 1: Build quintiles for both years
prescriber_volume_2022_q = df_2022.groupby('Prscrbr_NPI')['Tot_Clms'].sum().reset_index()
prescriber_volume_2022_q.columns = ['Prscrbr_NPI', 'total_glp1_claims']
prescriber_volume_2022_q['quintile'] = pd.qcut(
    prescriber_volume_2022_q['total_glp1_claims'],
    q=5,
    labels=range(1, 6),
    duplicates='drop'
)

prescriber_volume_2023_q = df.groupby('Prscrbr_NPI')['Tot_Clms'].sum().reset_index()
prescriber_volume_2023_q.columns = ['Prscrbr_NPI', 'total_glp1_claims']
prescriber_volume_2023_q['quintile'] = pd.qcut(
    prescriber_volume_2023_q['total_glp1_claims'],
    q=5,
    labels=range(1, 6),
    duplicates='drop'
)

# Step 2: Merge quintile labels onto each year's data
df_2022_q = df_2022.merge(
    prescriber_volume_2022_q[['Prscrbr_NPI', 'quintile']],
    on='Prscrbr_NPI', how='left'
)
df_2023_q = df.merge(
    prescriber_volume_2023_q[['Prscrbr_NPI', 'quintile']],
    on='Prscrbr_NPI', how='left'
)

# Step 3: Brand share within each quintile, both years
def brand_share_by_quintile(dframe):
    grp = dframe.groupby(['quintile', 'Brnd_Name'], observed=True)['Tot_Clms'].sum().reset_index()
    grp['quintile_total'] = grp.groupby('quintile', observed=True)['Tot_Clms'].transform('sum')
    grp['share_pct'] = (grp['Tot_Clms'] / grp['quintile_total'] * 100).round(1)
    return grp.pivot(index='quintile', columns='Brnd_Name', values='share_pct').fillna(0)[
        ['Ozempic', 'Trulicity', 'Mounjaro', 'Rybelsus']
    ]

quintile_2022 = brand_share_by_quintile(df_2022_q)
quintile_2023 = brand_share_by_quintile(df_2023_q)

# Step 4: Compute YoY change at quintile level
quintile_change = quintile_2023 - quintile_2022

print("BRAND SHARE BY QUINTILE — 2022")
print("=" * 65)
print(quintile_2022.to_string())

print("\n" + "=" * 65)
print("BRAND SHARE BY QUINTILE — 2023")
print("=" * 65)
print(quintile_2023.to_string())

print("\n" + "=" * 65)
print("YoY CHANGE BY QUINTILE (2022 → 2023)")
print("=" * 65)
print(quintile_change.round(1).to_string())
```

```python
# ─────────────────────────────────────────────────────────────
# PRESSURE TEST 3: Year-over-year stability of three-segment structure
# Recompute within-specialty decile structure for 2022, compare to 2023
# ─────────────────────────────────────────────────────────────

top_5_specialties = [
    'Family Practice', 'Internal Medicine', 'Nurse Practitioner',
    'Endocrinology', 'Physician Assistant'
]

def specialty_decile_structure(dframe, year_label):
    """Build within-specialty deciles and return summary stats per specialty"""
    
    # Get per-prescriber volume + specialty
    prescriber_with_specialty = dframe.groupby(
        ['Prscrbr_NPI', 'Prscrbr_Type']
    )['Tot_Clms'].sum().reset_index()
    prescriber_with_specialty.columns = ['Prscrbr_NPI', 'Specialty', 'total_glp1_claims']
    
    # Filter to top 5
    prescriber_top5 = prescriber_with_specialty[
        prescriber_with_specialty['Specialty'].isin(top_5_specialties)
    ].copy()
    
    # Build within-specialty deciles
    prescriber_top5['within_specialty_decile'] = prescriber_top5.groupby(
        'Specialty'
    )['total_glp1_claims'].transform(
        lambda x: pd.qcut(x, q=10, labels=range(1, 11), duplicates='drop')
    )
    
    # Build summary: for each specialty, what does Decile 10 look like?
    summary_rows = []
    for spec in top_5_specialties:
        spec_data = prescriber_top5[prescriber_top5['Specialty'] == spec]
        d10 = spec_data[spec_data['within_specialty_decile'] == 10]
        d1 = spec_data[spec_data['within_specialty_decile'] == 1]
        
        summary_rows.append({
            'Specialty': spec,
            'Year': year_label,
            'Total_prescribers': len(spec_data),
            'D10_avg_claims': round(d10['total_glp1_claims'].mean(), 1),
            'D10_max_claims': int(d10['total_glp1_claims'].max()) if len(d10) > 0 else 0,
            'D1_avg_claims': round(d1['total_glp1_claims'].mean(), 1),
            'D10_to_D1_ratio': round(d10['total_glp1_claims'].mean() / d1['total_glp1_claims'].mean(), 1) if len(d1) > 0 and d1['total_glp1_claims'].mean() > 0 else 0
        })
    
    return pd.DataFrame(summary_rows)

# Run for both years
struct_2022 = specialty_decile_structure(df_2022, '2022')
struct_2023 = specialty_decile_structure(df, '2023')

print("WITHIN-SPECIALTY DECILE STRUCTURE — 2022")
print("=" * 75)
print(struct_2022.to_string(index=False))

print("\n" + "=" * 75)
print("WITHIN-SPECIALTY DECILE STRUCTURE — 2023")
print("=" * 75)
print(struct_2023.to_string(index=False))

# Compute the change in volume gap
print("\n" + "=" * 75)
print("YoY STABILITY CHECK: D10_avg_claims by specialty")
print("=" * 75)
comparison = pd.merge(
    struct_2022[['Specialty', 'D10_avg_claims']].rename(columns={'D10_avg_claims': 'D10_2022'}),
    struct_2023[['Specialty', 'D10_avg_claims']].rename(columns={'D10_avg_claims': 'D10_2023'}),
    on='Specialty'
)
comparison['change_pct'] = ((comparison['D10_2023'] - comparison['D10_2022']) / comparison['D10_2022'] * 100).round(1)
print(comparison.to_string(index=False))
```

```python
# ─────────────────────────────────────────────────────────────
# WHY-DRILL TEST: Hypothesis A — Clinical Evidence
# Question: Is Lilly's Decile 10 strength concentrated in endocrinology
# specifically, or spread across all specialties?
# ─────────────────────────────────────────────────────────────

# Step 1: Build (specialty, all-population decile, company) share table for 2023
# We use ALL-POPULATION deciles (not within-specialty) because the original
# split-pyramid finding was at the all-population decile level.

# Get decile labels onto the 2023 dataframe (you have this — df_2023_d)
# We need to also tag company

def assign_company(brand):
    if brand in ['Ozempic', 'Rybelsus']:
        return 'Novo'
    elif brand in ['Trulicity', 'Mounjaro']:
        return 'Lilly'
    return 'Other'

df_2023_d['Company'] = df_2023_d['Brnd_Name'].apply(assign_company)

# Top 5 specialties (96% of volume)
top_5_specialties = [
    'Family Practice', 'Internal Medicine', 'Nurse Practitioner',
    'Endocrinology', 'Physician Assistant'
]

# Filter to top 5 specialties
df_top5_2023 = df_2023_d[df_2023_d['Prscrbr_Type'].isin(top_5_specialties)].copy()

# Step 2: For each (specialty, decile), compute Novo vs Lilly share
specialty_decile_company = df_top5_2023.groupby(
    ['Prscrbr_Type', 'decile', 'Company'],
    observed=True
)['Tot_Clms'].sum().reset_index()

# Pivot to get Novo and Lilly side by side
specialty_decile_pivot = specialty_decile_company.pivot_table(
    index=['Prscrbr_Type', 'decile'],
    columns='Company',
    values='Tot_Clms',
    fill_value=0
).reset_index()

# Calculate Novo share within each (specialty, decile) cell
specialty_decile_pivot['novo_share_pct'] = (
    specialty_decile_pivot['Novo'] /
    (specialty_decile_pivot['Novo'] + specialty_decile_pivot['Lilly']) * 100
).round(1)

# Step 3: Pivot for readability — specialties as columns, deciles as rows
result = specialty_decile_pivot.pivot(
    index='decile',
    columns='Prscrbr_Type',
    values='novo_share_pct'
).fillna(0)

# Reorder columns
result = result[top_5_specialties]

print("NOVO SHARE BY (SPECIALTY x DECILE) — 2023")
print("=" * 80)
print(result.to_string())

# Step 4: Highlight Decile 10 specifically
print("\n" + "=" * 80)
print("DECILE 10 NOVO SHARE BY SPECIALTY")
print("=" * 80)
decile_10_view = result.loc[10].sort_values()
print(decile_10_view.to_string())
```
